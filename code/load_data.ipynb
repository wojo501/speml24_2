```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
# loading data
data_dir = '../data/cifar-10-batches-py'

classes = ('plane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck')
data = []
labels = []
batch_files = [f"data_batch_{i}" for i in range(1, 6)]
for batch_file in batch_files:
    try:
        full_path = os.path.join(data_dir, batch_file)
        batch = pd.read_pickle(full_path)
        for i in range(len(batch['labels'])):
            data.append(batch['data'][i])
            labels.append(batch['labels'][i])
    except:
        raise Exception(f"File {full_path} does not exist")
```

```python
# reshaping data
for i in range(len(data)):
    data[i] = data[i].reshape(3, 32, 32).transpose(1, 2, 0)
```

```python
# checking data
plt.imshow(data[11])
print(classes[labels[11]])
```
